## Introdução

Juros é a remuneração do dinheiro investido ou a taxa cobrada por um dinheiro pego emprestado. Queremos calcular o quanto você recebeu (montante) pelo quanto você investiu (capital). Ou o quanto você tem que pagar por uma dívida de muito tempo.

Para as perguntas a seguir, considere:

$M$ = Montante

$C$ = Capital inicial

$r$ = Taxa de juros

## Questão 1: Juros simples

Juros simples ($J_s$) são calculados sempre em relação ao capital inicial. Eles não mudam ao longo do tempo. Eles são calculados como:

Juros simples = Capital inicial × taxa de juros × número de períodos

$J_s = C \times r \times n$,

onde $n$ é o número de períodos, isto é, quantas vezes o juros incide sobre o valor.

Por exemplo, se estamos calculando a incidência em meses e queremos o valor final após 1 ano, são 12 meses e $n$ = 12.

A taxa $r$ deve ser um valor entre 0 e 1. Caso esteja em %, é necessário dividir o valor por 100.

Para saber o montante M (valor final), deve-se somar o valor dos juros ao capital inicial:

$ M = C + J_s$

que é equivalente a

$M = C + C \times r \times n$

ou

$M = C (1 + r \times n)$  **(Equação 1)**


Juros simples não variam ao longo do tempo. Se você começou pagando R$100 de juros por mês, esse será o valor até a quitação da dívida, independente do prazo.

## Questão 2) Juros compostos



A característica dos juros compostos é que, para calcular o rendimento após cada período, os juros serão contados em cima de C(1+r), ou seja, juros sobre juros. Não mais apenas de C como nos juros simples.


Suponha a seguinte situação: Um empréstimo de R$1000 reais foi realizado, com taxas de 10% ao ano.

Após o primeiro ano:

$R\$1,000.00 + (R\$1,000.00 × 10\%) = R\$1,100.00$

ou

$R\$1,000.00 (1 + 0.1) = R\$1,000.00 \times 1.1 = R\$1,100.00$


E após diversos anos?

interest-compound-flow.svg

Observe que os juros aumentam: 100, 110, 121...

Qual é a conta?

Segundo ano: $R\$1,000.00 + (R\$1,000.00 × 10\%)(R\$1,000.00 × 10\%)$

ou

Segundo ano: $R\$1,000.00 (1.1 \times 1.1)$

Terceiro ano: $R\$1,000.00 (1.1 \times 1.1 \times 1.1)$

E assim por diante

interest-compound-flow2.svg

Ao invés de escrever vários termos de multiplicação, é mais conveniente escrever como uma potência:

Após 5 anos: $R\$1,000 \times 1.10^5 = R\$1,610.51$

E assim por diante.

Em resumo, uma fórmula para representar os juros compostos é:

$M = C (1+r)^t$ **(Equação 2)**

onde t é o número de anos (períodos) sendo considerado.

Observe como ela é diferente dos juros simples, pois há "juros sobre juros"

```python
# importei as bibliotecas para gerar um gráfico
import numpy as np
import matplotlib.pyplot as plt

# 2a)
def juros_composto(capital, taxa, tempo):
    return capital * (1 + taxa)**tempo # montando o gráfico de acordo com o enunciado

# 2b)
print(f'Montante após 10 anos: R$ {juros_composto(1000, 0.15, 10):.2f}') # alocando os valores pedidos

# 2c)
# definindo as arrays
anos = np.arange(1, 11)
montante = [juros_composto(1000, 0.15, t) for t in anos]

# gerando o gráfico
plt.plot(anos, montante, color='b')
plt.title('montante acumulado com juros composto (15% ao ano)')
plt.xlabel('ano')
plt.ylabel('montante (R$)')
plt.grid()
plt.xticks(anos)
plt.show()
```

## Questão 3) Cobrança cada vez mais rápida de juros


É possível tornar a fórmula para juros compostos ainda mais geral.

Na questão 2, a cobrança era anual.
Suponha que o objetivo seja calcular os juros dentro de intervalos menores do período. Por exemplo, queremos calcular o montante ao final de 1 ano. No entanto, agora a a incidência de juros acontece em intervalos menores:
- semestral
- trimestral
- bimestral
- mensal
- semanal
- diária
- ...

Se a gente variar a incidência de juros dessas diferentes maneiras, vocês acham que o valor final a pagar será sempre o mesmo? Vamos verificar.

Aproveitando os símbolos da **Equação 2**, outra expressão para juros compostos é:

$M=C\left(1+{\frac {r}{n}}\right)^{tn}$ **(Equação 3**)

onde :
- $M$ é o montante final
- $P$ é o capital inicial
- $t$ é o tempo em que os juros são aplicados. Por exemplo, aqui consideraremos ele como anos.
- $r$ é a taxa dos juros dentro do período $t$. Ou seja, consideraremos aqui a taxa de juros anual (mesmo tempo que $t$).
- $n$ é a frequência de composição. Em outras palavras, é o número de vezes por ano em que há incidência de juros
 - $n$ = 1: anual
 - $n$ = 12: mensal
 - $n$ = 52: semanal
 - $n$ = 365: diário

### Perguntas:

3a) Traduza a nova fórmula de juros compostos **(Equação 3)** para uma função de Python

3b) Suponha que o Capital inicial $C$ seja de apenas um real, isto é, R\$1.00. E que a taxa de juros $r$ seja de $\mathbf{100\%}$ (lembre-se de converter para decimal)

No período de 1 ano (t = 1), calcule o montante final para cada $n$ indicado:

 - $n$ = 1: anual
 - $n$ = 12: mensal
 - $n$ = 52: semanal
 - $n$ = 365: diário


 Uma calculadora de juros compostos está disponível em:

Utilize-a para comparar com os seus resultados


 3c) O que está acontecendo com o valor, na medida que a incidência de juros é mais rápida?



```python
# 3a)
def juros_composto(capital, taxa, tempo, frequencia):
    return capital * (1 + (taxa / frequencia))**(tempo * frequencia) # traduzindo o gráfico

# 3b)
print(f'Montante final em 1 ano: R$ {juros_composto(1.00, 1.0, 1, 1):.2f}\n'
      f'Montante final em 12 meses: R$ {juros_composto(1.00, 1.0, 1, 12):.2f}\n'
      f'Montante final em 52 semanas: R$ {juros_composto(1.00, 1.0, 1, 52):.2f}\n'
      f'Montante final em 365 dias: R$ {juros_composto(1.00, 1.0, 1, 365):.2f}')

# 3c)
"""
Conforme a frequência de capitalização aumenta, o valor final também aumenta.
Isso acontece pois os juros são aplicados mais frequentemente ao longo do tempo, aumentando o efeito dos juros sobre juros.
"""
```

### 3c)

Conforme a frequência de capitalização aumenta, o valor final também aumenta.
Isso acontece pois os juros são aplicados mais frequentemente ao longo do tempo, aumentando o efeito dos juros sobre juros.

## Questão 4) O que o número de Euler tem a ver com tudo isso?





O número de Euler pode ser visualizado através do Numpy utilizando a função numpy.e

O seu valor é 2.718281828459045

Na verdade, ela é um número irracional, com infinitas casas decimais que não se repetem.

Como podemos obtê-lo?

A sua origem é justamente na análise de juros compostos. Na medida que a incidência de juros é mais rápida, o montante final aumenta. Mas será que aumenta sem parar? Existe um limite?

O limite é quando consideramos uma incidência "instantânea", isto é, quando utilizamos grandes valores de $n$.

```python
# importando numpy para pegar o número de euler e a função exp(1)
import numpy as np

# 4a)
def juros_composto(capital, taxa, tempo, frequencia):
    return capital * (1 + (taxa / frequencia))**(tempo * frequencia)

print(f'Resultado (n = 10000000): R$ {juros_composto(1.00, 1.0, 1, 10000000):.2f}\n')

# 4b) e 4c)
print(f' ---------- Comparação ---------- \n'
      f'Número de Euler: {np.e}\n'
      f'Função exp(1): {np.exp(1)}\n'
      f'Juros Composto: R$ {juros_composto(1.00, 1.0, 1, 10000000)}')
```
